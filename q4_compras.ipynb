# Questão 4 — Simulação de Carrinho de Compras

**Objetivo:** Calcular os preços finais com desconto e a economia total obtida.

**Regras de desconto:**
- Item < R$50 → 5% de desconto
- Item entre R$50 e R$150 (inclusive) → 10% de desconto
- Item > R$150 → 15% de desconto

```python
# Lista de preços originais
precos = [50, 120, 30, 200, 80, 15]
```

```python
# Tarefa 1: Criar lista com preços finais após desconto
precos_finais = []
descontos_aplicados = []

for preco in precos:
    if preco < 50:
        desconto = 0.05
    elif preco <= 150:
        desconto = 0.10
    else:
        desconto = 0.15
    descontos_aplicados.append(desconto)
    preco_final = preco * (1 - desconto)
    precos_finais.append(preco_final)

print("Detalhamento do carrinho:")
print(f"{'Item':<6} {'Preço Original':>16} {'Desconto':>10} {'Preço Final':>12}")
print("-" * 46)
for i, (orig, desc, final) in enumerate(zip(precos, descontos_aplicados, precos_finais)):
    print(f"  {i+1:<4} R$ {orig:>10.2f}   {desc*100:>5.0f}%   R$ {final:>8.2f}")
```

```python
# Tarefa 2: Total com desconto e economia total
total_original = 0
total_final = 0

for orig, final in zip(precos, precos_finais):
    total_original += orig
    total_final += final

economia_total = total_original - total_final

print(f"\nTotal original:         R$ {total_original:.2f}")
print(f"Total com desconto:     R$ {total_final:.2f}")
print(f"Economia total:         R$ {economia_total:.2f}")
```
