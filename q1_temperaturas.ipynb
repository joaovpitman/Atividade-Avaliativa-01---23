# Questão 1 — Classificação de Temperaturas

**Objetivo:** Classificar as temperaturas registradas durante a semana usando condicionais e estruturas de repetição.

**Regras:**
- Abaixo de 20°C → `"Frio"`
- Entre 20°C e 30°C (inclusive) → `"Agradável"`
- Acima de 30°C → `"Quente"`

```python
# Lista de temperaturas da semana
temperaturas = [18, 22, 30, 35, 28, 15, 40]
```

```python
# Tarefa 1: Criar lista com as classificações
classificacoes = []

for temp in temperaturas:
    if temp < 20:
        classificacoes.append("Frio")
    elif temp <= 30:
        classificacoes.append("Agradável")
    else:
        classificacoes.append("Quente")

print("Temperaturas e classificações:")
for i, (temp, classif) in enumerate(zip(temperaturas, classificacoes)):
    print(f"  Dia {i+1}: {temp}°C → {classif}")
```

```python
# Tarefa 2: Contar quantos dias houve em cada categoria
contagem = {"Frio": 0, "Agradável": 0, "Quente": 0}

for classif in classificacoes:
    contagem[classif] += 1

print("\nContagem por categoria:")
for categoria, qtd in contagem.items():
    print(f"  {categoria}: {qtd} dia(s)")
```
