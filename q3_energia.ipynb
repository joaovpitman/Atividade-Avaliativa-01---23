# Questão 3 — Monitoramento de Consumo de Energia

**Objetivo:** Monitorar o consumo energético diário, calcular estatísticas e emitir alertas.

**Regras:**
- Consumo < 100 kWh → `"Baixo"`
- Consumo entre 100 e 180 kWh (inclusive) → `"Moderado"`
- Consumo > 180 kWh → `"Alto"`

```python
# Consumo diário em kWh
consumo = [120, 95, 200, 180, 75, 220, 160]
```

```python
# Classificar cada dia e filtrar dias com consumo ALTO
classificacoes = []
dias_alto = []

for i, kwh in enumerate(consumo):
    if kwh < 100:
        classif = "Baixo"
    elif kwh <= 180:
        classif = "Moderado"
    else:
        classif = "Alto"
    classificacoes.append(classif)
    if classif == "Alto":
        dias_alto.append(i + 1)  # dia começa em 1

print("Consumo por dia:")
for i, (kwh, classif) in enumerate(zip(consumo, classificacoes)):
    print(f"  Dia {i+1}: {kwh} kWh → {classif}")

print(f"\nDias com consumo ALTO: Dia(s) {dias_alto}")
```

```python
# Tarefa 2: Total e média semanal
total_consumo = 0
for kwh in consumo:
    total_consumo += kwh

media_semanal = total_consumo / len(consumo)

print(f"Total de consumo da semana: {total_consumo} kWh")
print(f"Média semanal:              {media_semanal:.2f} kWh/dia")
```

```python
# Tarefa 3: Alerta se mais de 2 dias com consumo alto
if len(dias_alto) > 2:
    print(f"⚠️  ALERTA: {len(dias_alto)} dias com consumo ALTO detectados! Verifique o consumo energético.")
else:
    print(f"✅ Consumo sob controle. Apenas {len(dias_alto)} dia(s) com consumo alto.")
```
