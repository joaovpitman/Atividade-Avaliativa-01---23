# Questão 2 — Sistema de Avaliação de Alunos

**Objetivo:** Classificar as notas finais dos alunos e calcular estatísticas de aprovação.

**Regras:**
- Nota < 5 → `"Reprovado"`
- Nota entre 5 e 7 (exclusive) → `"Recuperação"`
- Nota >= 7 → `"Aprovado"`

```python
# Lista de notas finais
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]
```

```python
# Tarefa 1: Gerar lista de classificações
classificacoes = []

for nota in notas:
    if nota < 5:
        classificacoes.append("Reprovado")
    elif nota < 7:
        classificacoes.append("Recuperação")
    else:
        classificacoes.append("Aprovado")

print("Notas e situação dos alunos:")
for i, (nota, classif) in enumerate(zip(notas, classificacoes)):
    print(f"  Aluno {i+1}: Nota {nota:.1f} → {classif}")
```

```python
# Tarefa 2: Calcular estatísticas
total = len(notas)
aprovados = 0
reprovados = 0

for classif in classificacoes:
    if classif == "Aprovado":
        aprovados += 1
    elif classif == "Reprovado":
        reprovados += 1

percentual_reprovados = (reprovados / total) * 100

print("\nEstatísticas:")
print(f"  Total de alunos:       {total}")
print(f"  Quantidade de aprovados: {aprovados}")
print(f"  Percentual de reprovados: {percentual_reprovados:.1f}%")
```
